```python
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["0.7 % Tijorat maqsadida",
          "0.3 % O‘qish",
          "85.8 % Qarindoshlarni yo‘qlash",
          "8.3 % Bo‘sh vaqt va dam olish",
          "3.2 % Xizmat yuzasidan",
          "1.7 % Davolanish"]

data = [4, 2, 69, 14, 8, 5]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-15)

bbox_props = dict(boxstyle="square,pad=0.2", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.25*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Turistik maqsadlarda kelgan chet el fuqarolarining safar maqsadlari bo‘yicha taqsimlanishi, ming kishi")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['2015-y', '2016-y', '2017-y', '2018-y', '2019-y', '2020-y', '2021-y']
men_means = [861.6, 898.7, 1193.5, 2515.8, 3218.5, 765.4, 956.7]
women_means = [1056.1, 1128.3, 1496.5, 2830.5, 3530.0, 738.7, 924.6]

x = np.arange(len(labels))  # the label locations
width = 0.40  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='erkaklar')
rects2 = ax.bar(x + width/2, women_means, width, label='ayollar')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('kelgan chet el fuqarolari')
ax.set_title('O‘zbekiston Respublikasiga kelgan chet el fuqarolarining davlatlar bo‘yicha taqsimlanishi, ming kishi')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
import plotly.express as px
import pandas as pd
stages = ["Qarindoshlarni yo'qlash", "Bo‘sh vaqt va dam olish", "Xizmat yuzasidan", "Davolanish", "Tijorat maqsadida", "O‘qish"]
df_mtl = pd.DataFrame(dict(number=[769.1, 109.9, 51.3, 13.9, 7.6, 4.9], stage=stages))
df_mtl['office'] = 'erkaklar'
df_toronto = pd.DataFrame(dict(number=[844.0, 45.2, 9.4, 18.5, 6.1, 1.5], stage=stages))
df_toronto['office'] = 'ayollar'
df = pd.concat([df_mtl, df_toronto], axis=0)
fig = px.funnel(df, x='number', y='stage', color='office')
fig.show()
```
